```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model

canada=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")
plt.scatter(canada['year'],canada['per capita income (US$)'],color='crimson')


regression=linear_model.LinearRegression()
regression.fit(canada[['year']].values,canada['per capita income (US$)'])
df1=pd.DataFrame([[2017],[2018],[2019],[2020],[2021],[2022],[2023.6],[2024],[2025],[2026],[2027],[2028],[2029],[2030]])
y2=regression.predict(df1)
y2
```
